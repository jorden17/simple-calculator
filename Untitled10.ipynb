```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.histplot(data=tips, x='total_bill', bins=10)
plt.show()
```

```python
sns.boxplot(data=tips, x='day', y='total_bill')
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
flights = sns.load_dataset('flights')
flights_pivot = flights.pivot('month', 'year', 'passengers')
sns.heatmap(flights_pivot, annot=True, fmt="d")
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

```

```python
df = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [10, 15, 13, 17, 20]
})

fig = px.line(df, x='x', y='y', title='Simple Line Plot')
fig.show()
```

```python
fig = px.bar(df, x='x', y='y', title='Simple Bar Plot')
fig.show()
```

```python
data = px.data.tips()
fig = px.histogram(data, x='total_bill', nbins=20)
fig.show()
```

```python
fig = px.scatter(data, x='total_bill', y='tip', color='day', size='size', hover_data=['sex'])
fig.show()
```

```python
fig = px.scatter_3d(data, x='total_bill', y='tip', z='size', color='day')
fig.show()
```
